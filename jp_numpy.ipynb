布尔运算实例

```python
import numpy as np
import matplotlib.pyplot as plt
def mandelbrot(h,w,maxit=20,r=2):
    x=np.linspace(-2.5,1.5,4*h+1)
    y=np.linspace(-1.5,1.5,3*w+1)
    A,B=np.meshgrid(x,y)
    C=A+B*1j
    z=np.zeros_like(C)
    divtime=maxit+np.zeros(z.shape,dtype=int)
    
    for i in range(maxit):
        z=z**2+C
        diverge=abs(z)>r
        div_now=diverge&(divtime==maxit)
        divtime[div_now]=i
        z[diverge]=r
    return divtime
%config InlineBackend.figure_format='retina'#显示清晰
plt.imshow(mandelbrot(100,100))
plt.savefig('x-y-t.png',dpi=300,bbox_inches='tight')#保存清晰
```
