从sklearn中加载iris数据体

```python
import pandas as pd
from sklearn.datasets import load_iris
iris_datas=load_iris()#iris_data是一个数据体，包含data、target
# print(iris_datas.data)
iris_datas.data.shape
iris_datas.target
```

读出数据指定标签

```python
iris=pd.DataFrame(iris_datas.data,columns=['SepaLength','SepalWidth','PetalLength','PetalWidth'])
#SepaLengthCm: 花萼长度, 单位cm；SepalWidthCm: 花萼宽度, 单位cm；PetalLengthCm: 花瓣长度, 单位cm
# PetalWidthCm; 花瓣宽度, 单位cm
iris
```

apply筛选数据

```python
import pandas as pd 
import numpy as np
a=range(5)
b=range(5,10)
c=range(10,15)
data_1=pd.DataFrame([a,b,c]).T
data_1.columns=['a','b','c']
print('data_1',data_1)

data_1['values']=''
data_1['degree']=data_1['values']

def select(x):
    if 0<=x<2:
        return 1
    elif 2<=x<3:
        return 2
    elif 3<=x<4:
        return 3
    elif 4<=x<5:
        return 4
    elif 5<=x<6:
        return 5
data_1['values']=data_1['a'].apply(select)
data_1['degree']=np.sin(data_1['values'])
print(data_1.iloc[:,:])
data_1.loc[:,:]
```
