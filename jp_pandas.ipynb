```python
import numpy as np
import math
import pandas as pd

file=('D:/code/Python Code/data/temp/model_liangceng10Hz_0.2ms.dat')
data=pd.read_csv(file,header=None,sep='\t')
data.columns=['data']
data.info
data.iloc[:,:]
# data=open(file)
# data
data=data['data'].str.split(' ',expand=True)


def st(t,s, freqlow, freqhigh, alpha):
    TimeLen=len(t)
    dt=t[1]-t[0]
    nLevel=int((freqhigh-freqlow)/alpha)+1
    fre=np.linspace(freqlow,freqhigh,nLevel)
    wcoefs=np.zeros((nLevel,TimeLen),dtype=complex)
    temp = np.zeros((1, TimeLen),dtype=complex)
    sigma_f = np.power(fre,-1)
    for m in range(nLevel):
        f=fre[m]
        for n in range(TimeLen):
            Gauss_st= (1/(math.sqrt(2*math.pi)*sigma_f[m]))*np.exp(-0.5*(np.power(n*dt-t,2)/(sigma_f[m]*sigma_f[m])))*np.exp(-1.0j*2*math.pi*f*t)
            temp[0,n]=np.sum(np.dot(s, Gauss_st))*dt
        wcoefs[[m],:]=temp
    return wcoefs

t=np.arange(0,1,0.0002)
freqlow=1;
freqhigh=100;
s=data.iloc[:,0].astype(str).astype(float)

alpha=1
data_s=st(t,s, freqlow, freqhigh, alpha)

```
